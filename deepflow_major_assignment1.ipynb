```python
def analyze_missing_values(dataframe):
    missing_data = pd.DataFrame({
        'Column': dataframe.columns,
        'Missing_Count': dataframe.isnull().sum(),
        'Missing_Percentage': round(dataframe.isnull().sum() / len(dataframe) * 100, 2),
        'Data_Type': dataframe.dtypes
    }).sort_values('Missing_Percentage', ascending=False)

    missing_data = missing_data[missing_data['Missing_Count'] > 0]
    return missing_data

missing_analysis = analyze_missing_values(df)
print("Top 20 columns with missing values:")
print(missing_analysis.head(20))

# Categorize missing values by severity
high_missing = missing_analysis[missing_analysis['Missing_Percentage'] > 80]
medium_missing = missing_analysis[(missing_analysis['Missing_Percentage'] > 20) &
                                 (missing_analysis['Missing_Percentage'] <= 80)]
low_missing = missing_analysis[missing_analysis['Missing_Percentage'] <= 20]

print(f"\nMissing Value Categories:")
print(f" High missing (>80%): {len(high_missing)} columns")
print(f" Medium missing (20-80%): {len(medium_missing)} columns")
print(f" Low missing (20%): {len(low_missing)} columns")
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"Number of categorical columns: {len(categorical_cols)}")

print("\nTop categorical variables by unique values:")
cat_analysis = pd.DataFrame({
    'Column': categorical_cols,
    'Unique_Values': [df[col].nunique() for col in categorical_cols],
    'Missing_Percentage': [round(df[col].isnull().sum() / len(df) * 100, 2) for col in categorical_cols]
}).sort_values('Unique_Values', ascending=True)

print(cat_analysis)
```

```python
def find_column(df, possible_names):
    for name in possible_names:
        if name in df.columns:
            return name
    return None

sale_price_col = find_column(df, ['SalePrice', 'salePrice', 'price', 'Price'])
machine_hours_col = find_column(df, ['MachineHoursCurrentMeter', 'machineHoursCurrentMeter', 'MachineHours', 'Hours'])
year_made_col = find_column(df, ['YearMade', 'yearMade', 'Year_Made', 'year_made'])
state_col = find_column(df, ['state', 'State', 'STATE'])
product_group_col = find_column(df, ['ProductGroup', 'productGroup', 'Product_Group'])
```

```python
key_numerical = []
if sale_price_col:
    key_numerical.append(sale_price_col)
if year_made_col:
    key_numerical.append(year_made_col)
if machine_hours_col:
    key_numerical.append(machine_hours_col)

for col in key_numerical:
    if col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)][col]
        print(f"{col}: {len(outliers)} outliers ({(len(outliers)/len(df))*100:.1f}%)")
```

```python
current_year = 2012  # Based on the data, sales go up to 2012
if year_made_col:
    invalid_years = df[(df[year_made_col] < 1900) | (df[year_made_col] > current_year)]
    print(f" Invalid {year_made_col} values: {len(invalid_years)} records")
else:
    print(" Year made column not found")

# Check extreme machine hours
if machine_hours_col:
    extreme_hours = df[df[machine_hours_col] > 50000]  # More than ~25 years of operation
    print(f" Extreme machine hours (>50k): {len(extreme_hours)} records")

    # Check zero machine hours (might be missing)
    zero_hours = df[df[machine_hours_col] == 0]
    print(f" Zero machine hours: {len(zero_hours)} records")
else:
    print(" Machine hours column not found")
```

```python
df_clean = df.copy()

# Strategy 1: Drop columns with >95% missing values
high_missing_cols = missing_analysis[missing_analysis['Missing_Percentage'] > 80]['Column'].tolist()
print(f"Dropping {len(high_missing_cols)} columns with >95% missing values")
if high_missing_cols:
    df_clean = df_clean.drop(columns=high_missing_cols)
    print(f"Dropped columns: {high_missing_cols[:5]}{'...' if len(high_missing_cols) > 5 else ''}")

if year_made_col:
    print(f"Fixing {year_made_col} anomalies...")
    df_clean[year_made_col] = df_clean[year_made_col].clip(lower=1950, upper=2012)
    print(f" {year_made_col} values capped to reasonable range (1950-2012)")

# Handle extreme machine hours
if machine_hours_col:
    print("Handling extreme machine hours...")
    # Cap machine hours at 99th percentile to remove extreme outliers
    hours_99th = df_clean[machine_hours_col].quantile(0.99)
    df_clean[machine_hours_col] = df_clean[machine_hours_col].clip(upper=hours_99th)
    print(f" Machine hours capped at 99th percentile ({hours_99th:.0f} hours)")

    # Convert zero hours to NaN (likely missing data)
    df_clean.loc[df_clean[machine_hours_col] == 0, machine_hours_col] = np.nan
    print(" Zero hours converted to NaN")
else:
    print("  Machine hours column not found - skipping hours processing")
    hours_99th = None

# 2.2 Feature Engineering from Sale Date
print("\n2.2 FEATURE ENGINEERING FROM SALE DATE")
print("-" * 40)

# Find sale date columns
sale_year_col = find_column(df, ['saleYear', 'sale_year', 'SaleYear', 'year'])
sale_month_col = find_column(df, ['saleMonth', 'sale_month', 'SaleMonth', 'month'])

# Equipment age at time of sale
if year_made_col and sale_year_col:
    df_clean['EquipmentAge'] = df_clean[sale_year_col] - df_clean[year_made_col]
    print(" Created EquipmentAge feature")
else:
    print("  Cannot create EquipmentAge - missing year columns")

# Sale season
if sale_month_col:
    def get_season(month):
        if month in [12, 1, 2]:
            return 'Winter'
        elif month in [3, 4, 5]:
            return 'Spring'
        elif month in [6, 7, 8]:
            return 'Summer'
        else:
            return 'Fall'

    df_clean['SaleSeason'] = df_clean[sale_month_col].apply(get_season)
    print(" Created SaleSeason feature")
else:
    print("  Cannot create SaleSeason - missing month column")

# Hours per year (usage intensity)
if machine_hours_col and 'EquipmentAge' in df_clean.columns:
    df_clean['HoursPerYear'] = df_clean[machine_hours_col] / (df_clean['EquipmentAge'] + 1)
    print(" Created HoursPerYear feature")
else:
    print("  Skipping HoursPerYear - missing required columns")

# 2.3 Handle Missing Values
print("\n2.3 HANDLING MISSING VALUES")
print("-" * 40)


# Strategy 2: Fill missing categorical values with 'Unknown'
categorical_cols_clean = df_clean.select_dtypes(include=['object']).columns.tolist()
for col in categorical_cols_clean:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna('Unknown')
print(f" Filled missing categorical values with 'Unknown'")

# Strategy 3: Handle specific important missing values
# Fill missing auctioneerID with mode
auctioneer_col = find_column(df_clean, ['auctioneerID', 'auctioneer_id', 'AuctioneerID'])
if auctioneer_col and auctioneer_col in df_clean.columns:
    df_clean[auctioneer_col] = df_clean[auctioneer_col].fillna(df_clean[auctioneer_col].mode()[0])
    print(f" Filled missing {auctioneer_col} with mode")

# Create missing indicator for machine hours
if machine_hours_col and machine_hours_col in df_clean.columns:
    df_clean['HasHoursMeter'] = (~df_clean[machine_hours_col].isna()).astype(int)

    # Fill missing machine hours with median by ProductGroup and EquipmentAge
    if product_group_col and 'EquipmentAge' in df_clean.columns:
        def fill_hours_by_group(group):
            return group[machine_hours_col].fillna(group[machine_hours_col].median())

else:
    # Create a dummy indicator if no hours column exists
    df_clean['HasHoursMeter'] = 0
    print("  No machine hours column found - created dummy HasHoursMeter feature")

# 2.4 Advanced Feature Engineering
print("\n2.4 ADVANCED FEATURE ENGINEERING")
print("-" * 40)

# Count non-null configuration features (equipment complexity)
id_cols = ['SalesID', 'MachineID', 'ModelID', 'datasource']
date_cols = [col for col in df_clean.columns if 'sale' in col.lower() or 'year' in col.lower() or 'month' in col.lower() or 'day' in col.lower()]
target_cols = [sale_price_col] if sale_price_col else []

exclude_cols = id_cols + date_cols + target_cols
config_cols = [col for col in df_clean.columns if col not in exclude_cols]

df_clean['ConfigurationCount'] = df_clean[config_cols].notna().sum(axis=1)
print(" Created ConfigurationCount feature (equipment complexity)")

# Price per hour (derived target for insights)
if machine_hours_col and machine_hours_col in df_clean.columns and sale_price_col:
    df_clean['PricePerHour'] = df_clean[sale_price_col] / (df_clean[machine_hours_col] + 1)
    print(" Created PricePerHour feature")

# Equipment size categories
if sale_price_col:
    def categorize_equipment_size(price):
        if price < 20000:
            return 'Small'
        elif price < 50000:
            return 'Medium'
        elif price < 100000:
            return 'Large'
        else:
            return 'Extra_Large'

    df_clean['EquipmentSizeCategory'] = df_clean[sale_price_col].apply(categorize_equipment_size)
    print(" Created EquipmentSizeCategory feature")

# 2.5 Handle Categorical Encoding
print("\n2.5 CATEGORICAL VARIABLE ENCODING")
print("-" * 40)

# Identify categorical columns for encoding
categorical_for_encoding = df_clean.select_dtypes(include=['object']).columns.tolist()

# Remove target and ID columns
id_and_target_cols = ['SalesID', 'MachineID', 'ModelID']
if sale_price_col:
    id_and_target_cols.append(sale_price_col)

categorical_for_encoding = [col for col in categorical_for_encoding
                           if col not in id_and_target_cols]

print(f"Categorical columns to encode: {len(categorical_for_encoding)}")

# Strategy: Use label encoding for high cardinality, one-hot for low cardinality
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
high_cardinality_threshold = 50

for col in categorical_for_encoding:
    if col in df_clean.columns:
        unique_count = df_clean[col].nunique()

        if unique_count > high_cardinality_threshold:
            # Use label encoding for high cardinality
            le = LabelEncoder()
            df_clean[f'{col}_encoded'] = le.fit_transform(df_clean[col].astype(str))
            label_encoders[col] = le
            df_clean = df_clean.drop(columns=[col])
            print(f" Label encoded {col} ({unique_count} unique values)")
        else:
            # Use one-hot encoding for low cardinality
            dummies = pd.get_dummies(df_clean[col], prefix=col, prefix_sep='_')
            df_clean = pd.concat([df_clean, dummies], axis=1)
            df_clean = df_clean.drop(columns=[col])
            print(f" One-hot encoded {col} ({unique_count} unique values)")

# 2.6 Final Data Types Correction
print("\n2.6 FINAL DATA TYPES CORRECTION")
print("-" * 40)

# Ensure all numeric columns are proper numeric types
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
for col in numeric_cols:
    if sale_price_col and col != sale_price_col:  # Keep target as float
        if df_clean[col].dtype == 'float64' and df_clean[col].nunique() < 1000:
            df_clean[col] = pd.to_numeric(df_clean[col], downcast='integer', errors='ignore')

print(" Optimized data types for memory efficiency")

# 2.7 Remove Irrelevant Columns
print("\n2.7 REMOVING IRRELEVANT COLUMNS")
print("-" * 40)

# Remove columns that won't help prediction
columns_to_drop = ['SalesID', 'MachineID']  # Keep ModelID as it might be useful
existing_cols_to_drop = [col for col in columns_to_drop if col in df_clean.columns]
if existing_cols_to_drop:
    df_clean = df_clean.drop(columns=existing_cols_to_drop)
    print(f" Dropped irrelevant columns: {existing_cols_to_drop}")

# 2.8 Final Dataset Summary
print("\n2.8 FINAL CLEANED DATASET SUMMARY")
print("-" * 40)

print(f"Final dataset shape: {df_clean.shape}")
print(f"Columns removed: {df.shape[1] - df_clean.shape[1]}")
print(f"Memory usage: {df_clean.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

# Check for remaining missing values
remaining_missing = df_clean.isnull().sum().sum()
print(f"Remaining missing values: {remaining_missing}")

if remaining_missing > 0:
    print("Columns with remaining missing values:")
    missing_cols = df_clean.isnull().sum()[df_clean.isnull().sum() > 0]
    print(missing_cols)

# 2.9 Feature Summary
print("\n2.9 FEATURE SUMMARY")
print("-" * 40)

feature_types = {
    'Numeric': len(df_clean.select_dtypes(include=[np.number]).columns),
    'Binary': len([col for col in df_clean.columns if df_clean[col].nunique() == 2]),
    'Total Features': df_clean.shape[1] - (1 if sale_price_col in df_clean.columns else 0)
}

for feature_type, count in feature_types.items():
    print(f" {feature_type}: {count}")

# 2.10 Save Cleaned Dataset
print("\n2.10 SAVING CLEANED DATASET")
print("-" * 40)

# Save the cleaned dataset
df_clean.to_csv('train_cleaned.csv', index=False)
print(" Cleaned dataset saved as 'train_cleaned.csv'")
```
